## Aggregate all VOTT output

```python
import os
import shutil
from xml.etree import ElementTree as et

# Create Annotations Folder
annotation_output_dir = os.path.join(output_dir, 'Annotations');
if not os.path.isdir(annotation_output_dir):
    os.makedirs(annotation_output_dir)

# Create JPEGImages Folder
jpeg_output_dir = os.path.join(output_dir, 'JPEGImages');
if not os.path.isdir(jpeg_output_dir):
    os.makedirs(jpeg_output_dir)
    
# Create ImageSets Folder
main_output_dir = os.path.join(output_dir, 'ImageSets', 'Main');
if not os.path.isdir(main_output_dir):
    os.makedirs(main_output_dir)
    
train_output_file = os.path.join(main_output_dir,'Fish_train.txt')
val_output_file = os.path.join(main_output_dir,'Fish_val.txt')
    
tf_out = open(train_output_file, 'w')
vf_out = open(val_output_file, 'w')

total_count_train = 0 
total_count_test = 0

for directory in os.listdir(source_dir):
    if directory[-6:] == 'output':

        annotation_dir = os.path.join(source_dir, directory, 'TensorFlow', 'Annotations')
        jpeg_dir = os.path.join(source_dir, directory, 'TensorFlow', 'JPEGImages')
        main_dir = os.path.join(source_dir, directory, 'TensorFlow', 'ImageSets', 'Main')

        count_files = 0
        # Copying the .xml files
        for file in sorted(os.listdir(annotation_dir)): 
            data_file = os.path.join(annotation_dir,file)
            # Just to ensure the folder contains only .xml files
            if (data_file[-4:] == '.xml'):
                basename = file[:-4]
                newname = basename + "_" + directory
                new_xml_name = os.path.join(annotation_output_dir, newname + '.xml')
                new_jpg_name = os.path.join(jpeg_output_dir, newname + '.jpg')

                # Copy the file
                shutil.copy(data_file, new_xml_name)

                # Open it and update filename
                tree = et.parse(new_xml_name)
                tree.find('filename').text = newname
                tree.find('path').text = new_jpg_name
                tree.write(new_xml_name)
                count_files += 1

        print("copied {} xml files from {}".format(count_files, directory))

        count_files = 0
        # Copying the .jpg files
        for file in sorted(os.listdir(jpeg_dir)):
            jpeg_file = os.path.join(jpeg_dir,file)
            # Just to ensure the folder contains only .jpg files
            if (jpeg_file[-4:] == '.jpg'):
                basename = file[:-4]
                new_jpg_name = os.path.join(jpeg_output_dir, basename + "_" + directory + '.jpg')

                # Copy the file
                shutil.copy(jpeg_file, new_jpg_name)
                count_files += 1

        print("copied {} jpg files from {}".format(count_files, directory))

        train_file = os.path.join(main_dir,'Fish_train.txt')

        with open(train_file) as tf:
            content = tf.readlines()
            count_files = 0

            for line in content :
                name = value = line[:-1].split()[0] + "_" + directory
                value = line[:-1].split()[1]

                line = name + " " + value + "\n"
                tf_out.write(line)
                count_files += 1
                total_count_train += 1

        print("adding {} lines in {}".format(count_files, 'Fish_train.txt'))

        val_file = os.path.join(main_dir,'Fish_val.txt')

        with open(val_file) as vf:
            content = vf.readlines()
            count_files = 0

            for line in content :
                name = value = line[:-1].split()[0] + "_" + directory
                value = line[:-1].split()[1]

                line = name + " " + value + "\n"
                vf_out.write(line)
                count_files += 1
                total_count_test += 1

        print("adding {} lines in {}".format(count_files, 'Fish_val.txt'))
        
vf_out.close()
tf_out.close()

print("total count of training samples: {}".format(total_count_train))
print("total count of testing samples: {}".format(total_count_test))

```
